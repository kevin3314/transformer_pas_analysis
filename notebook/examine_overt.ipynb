```python
from pathlib import Path
from kyoto_reader import KyotoReader, Document, Argument, SpecialArgument
```

```python
reader = KyotoReader(Path('/Users/NobuhiroUeda/Data/kwdlc/train'))
```

```python
docs = list(reader.process_all_documents())
```

```python
def examine(doc):
    cnt = 0
    for bp in doc.bp_list():
        feats = bp.tag.features
        for case in ('ガ', 'ヲ', 'ニ', 'ガ２', 'ノ', 'ノ？', '判ガ'):
            old = (case in ('ノ', 'ノ？') and feats.get('係', None) in ('ノ格', 'ノ？格')) or case.lstrip('判') in feats
            dep_case = feats.get('係', '').rstrip('格')
            new = (case == dep_case) or (case == '判ガ' and dep_case == 'ガ') or (case == 'ノ？' and dep_case == 'ノ')
            if old is not new:
                import ipdb; ipdb.set_trace()
                cnt += 1
    return cnt
```

```python
def examine_pred(doc):
    cnt = 0
    for pred in doc.get_predicates():
        for case, args in doc.get_arguments(pred).items():
            if case not in ('ガ', 'ヲ', 'ニ', 'ガ２', 'ノ', 'ノ？', '判ガ'):
                continue
            for arg in args:
                if not isinstance(arg, Argument):
                    continue
                feats = arg.tag.features
                old = (case in ('ノ', 'ノ？') and feats.get('係', None) in ('ノ格', 'ノ？格')) or case.lstrip('判') in feats
                dep_case = feats.get('係', '').rstrip('格')
                new = (case == dep_case) or (case == '判ガ' and dep_case == 'ガ') or (case == 'ノ？' and dep_case == 'ノ')
                if old is True and new is False:
                    print(doc.doc_id, pred.tag.midasi, case, arg.tag.midasi)
                    cnt += 1
    return cnt

for doc in docs:
#     print(doc.doc_id, examine_pred(doc))
    examine_pred(doc)
```

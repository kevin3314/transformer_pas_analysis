```python
from pathlib import Path
from kyoto_reader import KyotoReader, Document, Argument, SpecialArgument
```

```python
reader = KyotoReader('/Users/NobuhiroUeda/Data/kwdlc/train')
```

```python
docs = list(reader.process_all_documents())
```

```python
# 述語に対する rel がアノテーションされているのにも関わらず，述語判定されない例を探す
n_has_arg = 0
n_is_pred = 0
n_both = 0

def examine(doc):
    global n_has_arg, n_is_pred, n_both
    for pred in doc.get_predicates():
        arguments = doc.get_arguments(pred)
        has_arg = any(arguments[case] for case in ('ガ', 'ヲ', 'ニ', 'ガ２'))
        is_pred = '用言' in pred.tag.features or '非用言格解析' in pred.tag.features
        if has_arg and not is_pred:
            print(doc[pred.sid], pred.midasi)
            print(pred.tag.features)
        if has_arg:
            n_has_arg += 1
        if is_pred:
            n_is_pred += 1
        if has_arg and is_pred:
            n_both += 1

for doc in docs:
    examine(doc)

n_has_arg, n_is_pred, n_both
```
